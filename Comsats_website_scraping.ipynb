```python
import requests
from bs4 import BeautifulSoup

r = requests.get('https://lahore.comsats.edu.pk/default.aspx')
soup = BeautifulSoup(r.content,'html.parser')
```

```python
admissions = soup.find_all('p')[1].text
faculty_and_department = soup.find_all('p')[2].text
research = soup.find_all('p')[3].text
scholarships = soup.find_all('p')[4].text

print('1) Topics in Comsats lahore website are:')
print('2) Admissions:',admissions)
print('3) Faculty & Department:',faculty_and_department)
print('4) Research:',research)
print('5) Scholarships:',scholarships)
```
