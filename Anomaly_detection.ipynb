```python
print('Mean value of cgpa:',df['cgpa'].mean())
print('std value of cgpa:',df['cgpa'].std())
print('Min value of cgpa:',df['cgpa'].min())
print('Max value of cgpa:',df['cgpa'].max())
```

```python
#Print boundaries
print('Highest allowed:',df['cgpa'].mean() + 3*df['cgpa'].std())
print('Lowest allowed:',df['cgpa'].mean() - 3*df['cgpa'].std())
```

```python
df[(df['cgpa'] > 8.80) | (df['cgpa'] < 5.11)]
```

```python
#Trimming
upper_bound = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_bound = df['cgpa'].mean() - 3*df['cgpa'].std()
new_df = df[(df['cgpa'] < upper_bound) & (df['cgpa'] > lower_bound)]
```

```python
new_df
```

```python
#Using z-score method
df['z_score_cgpa'] = (df['cgpa'] - df['cgpa'].mean())/df['cgpa'].std()
df
```

```python
df_new = df[(df['z_score_cgpa']<3) & (df['z_score_cgpa']>-3)]
df_new
```

```python
df['cgpa'] = np.where((df['cgpa'] > upper_bound),upper_bound,np.where((df['cgpa'] < lower_bound),lower_bound,df['cgpa']))
```
